```python
import pickle
from IPython.display import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
functionclass = "linear regression"
dataset = "celeba"
k = 30

with open('./alex/celeba_mmr_{}_linearregression.pkl'.format(k), 'rb+') as f: 
        results_mmr_30_lr = pickle.load(f)

with open('./alex/celeba_oracle_{}_linearregression.pkl'.format(k), 'rb+') as f: 
        results_oracle_30_lr = pickle.load(f)


print(results_mmr_30_lr.keys())
print(results_oracle_30_lr.keys())

sim_baseline = results_mmr_30_lr['sims'][0]
rep_baseline = results_mmr_30_lr['MPR'][0]

plt.figure()
plt.title("Similarity Over MPR, {}, {}, k={}".format(functionclass, dataset,k))
plt.plot(results_oracle_30_lr["rounded_MPR"], results_oracle_30_lr["rounded_sims"], label="Binary")
plt.plot(results_oracle_30_lr["relaxed_MPR"], results_oracle_30_lr["relaxed_sims"], label="Relaxed")
plt.plot(results_mmr_30_lr["MPR"], results_mmr_30_lr["sims"], label="MMR")
plt.axhline(y=sim_baseline, color='b', linestyle=':', label="Baseline KNN Similarity")
plt.axvline(x=rep_baseline, color='r', linestyle=':', label="Baseline KNN MPR")
plt.xlabel('MPR')
plt.ylabel('Similarity')
plt.legend()
plt.savefig("Similarity Over MPR, {}, {} K={}.png".format(functionclass, dataset,k))
```

```python
functionclass = "linear regression"
dataset = "celeba"
k = 30

with open('./alex/celeba_mmr_{}_linearregression.pkl'.format(k), 'rb+') as f: 
        results_mmr_30_lr = pickle.load(f)

with open('./alex/celeba_oracle_{}_linearregression.pkl'.format(k), 'rb+') as f: 
        results_oracle_30_lr = pickle.load(f)


print(results_mmr_30_lr.keys())
print(results_oracle_30_lr.keys())


k2 = 10
with open('./alex/celeba_mmr_{}_linearregression.pkl'.format(k2), 'rb+') as f: 
        results_mmr_10_lr = pickle.load(f)

with open('./alex/celeba_oracle_{}_linearregression.pkl'.format(k2), 'rb+') as f: 
        results_oracle_10_lr = pickle.load(f)

results_oracle_30_lr["rounded_sims"] = np.array(results_oracle_30_lr["rounded_sims"])/30
results_oracle_30_lr["relaxed_sims"] = np.array(results_oracle_30_lr["relaxed_sims"])/30
results_mmr_30_lr["sims"] = np.array(results_mmr_30_lr["sims"])/30
results_oracle_10_lr["rounded_sims"] = np.array(results_oracle_10_lr["rounded_sims"])/10
results_oracle_10_lr["relaxed_sims"] = np.array(results_oracle_10_lr["relaxed_sims"])/10
results_mmr_10_lr["sims"] = np.array(results_mmr_10_lr["sims"])/10


plt.figure()
plt.title("Similarity Over MPR, {}, {}, L={},{}".format(functionclass, dataset,k,k2))
plt.plot(results_oracle_30_lr["rounded_MPR"], results_oracle_30_lr["rounded_sims"], label="Binary, K={}".format(k))
plt.plot(results_oracle_30_lr["relaxed_MPR"], results_oracle_30_lr["relaxed_sims"], label="Relaxed, K={}".format(k))
plt.plot(results_mmr_30_lr["MPR"], results_mmr_30_lr["sims"], label="MMR, K={}".format(k))

plt.plot(results_oracle_10_lr["rounded_MPR"], results_oracle_10_lr["rounded_sims"], label="Binary, K={}".format(k2))
plt.plot(results_oracle_10_lr["relaxed_MPR"], results_oracle_10_lr["relaxed_sims"], label="Relaxed, K={}".format(k2))
plt.plot(results_mmr_10_lr["MPR"], results_mmr_10_lr["sims"], label="MMR")
# plt.axhline(y=sim_baseline, color='b', linestyle=':', label="Baseline KNN Similarity")
# plt.axvline(x=rep_baseline, color='r', linestyle=':', label="Baseline KNN MPR")
plt.xlabel('MPR')
plt.ylabel('Mean Similarity')
plt.legend()
plt.savefig("Similarity Over MPR, {}, {} K={}.png".format(functionclass, dataset,k))
```
